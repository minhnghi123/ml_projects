```python
df['Which tools or technologies have you worked with? '].value_counts()
# đếm xem có bao nhiêu người sử dụng tools và công nghệ tương ứng
```

```python
df['Which tools or technologies have you worked with? '] = (
    df['Which tools or technologies have you worked with? ']
    .str.strip()
    .str.replace(r'\s*,\s*', ',', regex=True)  # loại bỏ khoảng trắng quanh dấu phẩy
)
# Bước xử lý clean dữ liệu , xóa khoảng trắng hai đầu, khoảng trăng xung quanh dấu phải
```

```python
tools_series = df['Which tools or technologies have you worked with? '].dropna().str.split(',')
#tiêp theo là drop các missing values và split(',') loại bỏ dấu phẩy.
```

```python
tools_series
```

```python
from itertools import chain
flat_tools = list(chain.from_iterable(tools_series))
# sử dụng chain để nối các chuỗi các series lại thành một list duy nhất .
```

```python
flat_tools
```

```python
tool_counts = pd.Series(flat_tools).value_counts()
print(tool_counts)
# tiếp theo là chuyển đổi list đó về thành series rồi valeus_count() để đếm có bao nhiêu người  dùng tools tương ứng
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=tool_counts.values[:10], y=tool_counts.index[:10], palette='crest')
plt.title('Top 10 Tools & Technologies Used')
plt.xlabel('Number of People')
plt.ylabel('Tool / Technology')
plt.tight_layout()
plt.show() ;
#Cuối cùng vẽ sơ đồ , dùng barplot và hiện thị 10 features nổi nhất.
```
